## 1. Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
files = {
    'Issue Frequency Trends': 'System Analysis Sheets - 1. Issue Frequency Trends.csv',
    'Critical System Issues': 'System Analysis Sheets - 2. Critical System Issues.csv',
    'Server Maintenance': 'System Analysis Sheets - 3.1 Server Maintenance.csv',
    'Storage System Status': 'System Analysis Sheets - 3.2 Storage System Status.csv',
    'Patch History': 'System Analysis Sheets - 4.1 Patch History.csv',
    'Database Performance History': 'System Analysis Sheets - 5.1 Database Performance History.csv',
    'Critical Interventions': 'System Analysis Sheets - 6.1 Critical Interventions.csv',
    'System Resource Usage': 'System Analysis Sheets - 7.1 System Resource Usage.csv'
}

# Read CSV files into DataFrames
dataframes = {name: pd.read_csv(f'./{filename}') for name, filename in files.items()}

# Display first rows of each dataset
{key: df.head() for key, df in dataframes.items()}
```

## 2. Issue Trends Over Time

```python
df_trends = dataframes['Issue Frequency Trends']
plt.figure(figsize=(10, 5))
for col in ['Critical Issues', 'Major Issues', 'Minor Issues', 'Emergency Fixes']:
    plt.plot(df_trends['Year'], df_trends[col], marker='o', label=col)
plt.xlabel('Year')
plt.ylabel('Number of Issues')
plt.title('Issue Trends Over Time')
plt.legend()
plt.grid()
plt.show()
```

## 3. Critical Failures & Downtime

```python
df_issues = dataframes['Critical System Issues']
plt.figure(figsize=(10, 5))
sns.barplot(x=df_issues['Issue'], y=df_issues['Downtime'])
plt.xlabel('Issue Type')
plt.ylabel('Downtime (hours)')
plt.title('Downtime by Critical Issue')
plt.xticks(rotation=45)
plt.show()
```

## 4. Server & Storage Health Check

```python
df_servers = dataframes['Server Maintenance']
plt.figure(figsize=(10, 5))
sns.heatmap(df_servers[['Replacement Due']].T, cmap='coolwarm', annot=True)
plt.title('Server Replacement Due Timeline')
plt.show()
```

## 5. Patch & Database Performance

```python
df_db = dataframes['Database Performance History']
plt.figure(figsize=(10, 5))
plt.plot(df_db['Year'], df_db['Query Time'], marker='o', label='Query Time')
plt.plot(df_db['Year'], df_db['Index Fragmentation'], marker='s', label='Index Fragmentation')
plt.xlabel('Year')
plt.ylabel('Performance Metrics')
plt.title('Database Query Performance Over Time')
plt.legend()
plt.grid()
plt.show()
```

## 6. Critical Interventions & System Resource Usage

```python
df_resources = dataframes['System Resource Usage']
plt.figure(figsize=(10, 5))
for col in ['CPU Peak', 'Memory Peak', 'Disk I/O', 'Network Usage']:
    plt.plot(df_resources['Year'], df_resources[col], marker='o', label=col)
plt.xlabel('Year')
plt.ylabel('Resource Usage')
plt.title('System Resource Consumption Trends')
plt.legend()
plt.grid()
plt.show()
```

## 7. Summary & Next Steps
- **Key Findings:** The system is aging, with increasing failures and high resource usage.
- **Recommended Actions:**
  - Upgrade failing servers & storage.
  - Improve backup & security measures.
  - Optimize database performance.
