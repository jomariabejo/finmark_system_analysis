## 1. Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
files = {
    'Issue Frequency Trends': 'System Analysis Sheets - 1. Issue Frequency Trends.csv',
    'Critical System Issues': 'System Analysis Sheets - 2. Critical System Issues.csv',
    'Server Maintenance': 'System Analysis Sheets - 3.1 Server Maintenance.csv',
    'Storage System Status': 'System Analysis Sheets - 3.2 Storage System Status.csv',
    'Patch History': 'System Analysis Sheets - 4.1 Patch History.csv',
    'Database Performance History': 'System Analysis Sheets - 5.1 Database Performance History.csv',
    'Critical Interventions': 'System Analysis Sheets - 6.1 Critical Interventions.csv',
    'System Resource Usage': 'System Analysis Sheets - 7.1 System Resource Usage.csv'
}

# Read CSV files into DataFrames
dataframes = {name: pd.read_csv(f'./{filename}') for name, filename in files.items()}

# Display first rows of each dataset
{key: df.head() for key, df in dataframes.items()}
```

## 2. Issue Trends Over Time

```python
import matplotlib.pyplot as plt

# Extracting the relevant DataFrame from the dictionary
# This DataFrame contains data on the frequency of different issue types over time
df_trends = dataframes['Issue Frequency Trends']

# Creating a figure with a defined size for better readability
plt.figure(figsize=(10, 5))

# Looping through different issue categories to plot their trends over the years
issue_types = ['Critical Issues', 'Major Issues', 'Minor Issues', 'Emergency Fixes']
for issue in issue_types:
    plt.plot(df_trends['Year'], df_trends[issue], marker='o', linestyle='-', label=issue)

# Adding labels and a title to the plot
plt.xlabel('Year')  # Label for x-axis representing the years
plt.ylabel('Number of Issues')  # Label for y-axis representing the issue count
plt.title('Issue Trends Over Time')  # Title of the plot

# Adding a legend to differentiate between issue types
plt.legend()

# Enabling grid for better visualization of data points
plt.grid(True)

# Displaying the plot
plt.show()

# This report visualizes the trends of different types of issues over the years.
# It helps identify whether the number of critical, major, minor issues, and emergency fixes is increasing or decreasing over time.
# A rising trend in critical or emergency fixes may indicate system instability and the need for better preventive measures.
# A downward trend suggests improvements in system stability and issue resolution strategies.
# This analysis aids in making informed decisions for IT infrastructure improvements and maintenance planning.
```

## 3. Critical Failures & Downtime

```python
# Extracting the DataFrame for critical system issues
df_issues = dataframes['Critical System Issues']

# Creating a bar plot to visualize downtime by issue type
plt.figure(figsize=(10, 5))
sns.barplot(x=df_issues['Issue'], y=df_issues['Downtime'])
plt.xlabel('Issue Type')  # Label for x-axis representing different issue types
plt.ylabel('Downtime (hours)')  # Label for y-axis representing downtime in hours
plt.title('Downtime by Critical Issue')  # Title of the plot
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.show()

# The bar chart illustrates the downtime associated with different critical system issues.
# This visualization helps identify which issues contribute the most to system downtime.
# Understanding these patterns allows IT teams to prioritize their resolution efforts efficiently.
# Organizations can use this data to implement preventive measures and minimize future disruptions.
```

## 4. Server & Storage Health Check

```python
# Extracting the DataFrame for server maintenance
df_servers = dataframes['Server Maintenance']

# Creating a heatmap to visualize server replacement due timeline
plt.figure(figsize=(10, 5))
sns.heatmap(df_servers[['Replacement Due']].T, cmap='coolwarm', annot=True)
plt.title('Server Replacement Due Timeline')
plt.show()

# The heatmap provides a clear visualization of when servers are due for replacement.
# This helps IT teams plan maintenance and budget for hardware upgrades efficiently.
# By identifying upcoming replacements, organizations can minimize unexpected failures and optimize resource allocation.
```

## 5. Patch & Database Performance

```python
# Extracting the DataFrame for database performance history
df_db = dataframes['Database Performance History']

# Creating a line plot to visualize query time and index fragmentation over time
plt.figure(figsize=(10, 5))
plt.plot(df_db['Year'], df_db['Query Time'], marker='o', label='Query Time')
plt.plot(df_db['Year'], df_db['Index Fragmentation'], marker='s', label='Index Fragmentation')
plt.xlabel('Year')  # Label for x-axis representing the years
plt.ylabel('Performance Metrics')  # Label for y-axis representing performance indicators
plt.title('Database Query Performance Over Time')  # Title of the plot
plt.legend()  # Adding legend to differentiate the lines
plt.grid()  # Enabling grid for better readability
plt.show()

# The line plot illustrates how database query time and index fragmentation have changed over the years.
# Rising query time or fragmentation may indicate performance degradation, requiring optimization strategies such as indexing or query tuning.
# This analysis helps database administrators make informed decisions to maintain optimal database performance.
```

## 6. Critical Interventions & System Resource Usage

```python
# Extracting the DataFrame for system resource usage
df_resources = dataframes['System Resource Usage']

# Creating a line plot to visualize system resource consumption trends
plt.figure(figsize=(10, 5))
for col in ['CPU Peak', 'Memory Peak', 'Disk I/O', 'Network Usage']:
    plt.plot(df_resources['Year'], df_resources[col], marker='o', label=col)
plt.xlabel('Year')  # Label for x-axis representing the years
plt.ylabel('Resource Usage')  # Label for y-axis representing system resource metrics
plt.title('System Resource Consumption Trends')  # Title of the plot
plt.legend()  # Adding legend to differentiate resource usage trends
plt.grid()  # Enabling grid for better readability
plt.show()

# The line plot illustrates trends in system resource consumption over the years.
# Monitoring CPU, memory, disk I/O, and network usage helps identify potential performance bottlenecks.
# If resource usage trends show excessive peaks, this may indicate areas where scaling or optimization is needed.
# Understanding these patterns enables IT teams to plan infrastructure improvements effectively.
# 
# Why is this important?
# Efficient resource management ensures smooth system operations, prevents slowdowns, and optimizes cost efficiency.
# Unchecked resource usage can lead to server crashes, degraded application performance, and increased operational costs.
# 
# Why do we need to improve it?
# Proactively addressing resource bottlenecks helps maintain system stability and performance, preventing costly downtime.
# Optimization efforts, such as load balancing or hardware upgrades, ensure scalability as business needs grow.
# 
# What issues could occur in the future if this is not resolved?
# - Increased system latency, negatively impacting user experience.
# - Higher risk of system failures or crashes, leading to operational disruptions.
# - Unnecessary infrastructure costs due to inefficient resource allocation.
# - Security vulnerabilities due to resource exhaustion, making systems more susceptible to attacks.
```
